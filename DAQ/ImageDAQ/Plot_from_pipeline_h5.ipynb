```python
import h5py as h5
import matplotlib.pyplot as plt
import numpy as np
```

## User inputs

```python
fN = 'SATFE10-PSCR0666.h5'
```

## Load data

```python
with h5.File(fN,'r') as fh:
    print(list(fh.keys()))
    IMG = fh['images'][:]
```

```python
IMG.shape
```

```python
axis = np.arange(-130,130,1)/35.9
plt.figure(figsize=[10,10])
plt.suptitle('SATFE10-PSCR066')
plt.subplot(221)
plt.pcolormesh(axis, axis, IMG[0,100:360,170:430])
plt.xlabel('[mm]')
plt.ylabel('[mm]')
plt.subplot(222)
plt.pcolormesh(axis,axis,IMG[25,100:360,170:430])
plt.xlabel('[mm]')
plt.ylabel('[mm]')
plt.savefig('PSCR066')
```
