```python
import numpy as np
import matplotlib.pyplot as plt
import epics as ep
```

## User inputs

```python
cameraString = 'SATFE10-PSCR066:'
numshots =10
```

## grab image and plot

```python
def getImg(cameraString,numshots):
    imagePV =cameraString+'FPICTURE'
    widthPV =cameraString+'WIDTH'
    heightPV =cameraString+'HEIGHT'

    width = int(ep.caget(widthPV))
    height = int(ep.caget(heightPV))
    
    Img = []
    for i in range(0,numshots):
        Img.append(ep.caget(imagePV).reshape([height, width]))
    return np.asarray(Img)
```

```python
test = getImg(cameraString,numshots)
```
