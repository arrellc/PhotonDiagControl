```python
import matplotlib.pyplot as plt
import epics as ep
from datetime import date
import h5py as h5
%matplotlib notebook
```

## User inputs

```python
image_1 = '/photonics/home/arrell_c/before_change.h5'
image_2 = '/photonics/home/arrell_c/after_change2.h5'
```

## Load and display

```python
with h5.File(image_1,'r') as fh:
    img_1 = fh['img'][:]
with h5.File(image_2,'r') as fh:
    img_2 = fh['img'][:]
```

```python
l_1 = [75,100]
l_2 = [135,160]
lineout_1 = img_1[l_1[0]:l_1[1],:].sum(axis=0)
lineout_2 = img_2[l_2[0]:l_2[1],:].sum(axis=0)
```

```python
plt.figure()

plt.subplot(221)
plt.title(image_1[25:])
plt.imshow(img_1)
lims = plt.xlim()
plt.hlines(l_1[0],lims[0],lims[1],color ='r', linestyles='dashed')
plt.hlines(l_1[1],lims[0],lims[1],color ='r', linestyles='dashed')

plt.subplot(222)
plt.title(image_2[25:])
plt.imshow(img_2)
lims = plt.xlim()
plt.hlines(l_2[0],lims[0],lims[1],color ='r', linestyles='dashed')
plt.hlines(l_2[1],lims[0],lims[1],color ='r', linestyles='dashed')

plt.subplot(223)
plt.plot(lineout_1/max(lineout_1))
plt.grid(True)

plt.subplot(224)
plt.plot(lineout_2/max(lineout_2))
plt.grid(True)

plt.tight_layout()
```
